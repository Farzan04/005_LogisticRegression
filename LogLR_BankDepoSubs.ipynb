```python
import pandas as pd
import numpy as np
```

```python
data_bank = pd.read_csv('bank-full.csv', sep=';',quotechar='"')
data_bank
```

```python
data_bank.shape
```

```python
data_bank.info()
```

```python
data_bank.describe()
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
label = LabelEncoder()
data_bank['job'] = label.fit_transform(data_bank['job'])
data_bank['marital'] = label.fit_transform(data_bank['marital'])
data_bank['education'] = label.fit_transform(data_bank['education'])
data_bank['default'] = label.fit_transform(data_bank['default'])
data_bank['housing'] = label.fit_transform(data_bank['housing'])
data_bank['loan'] = label.fit_transform(data_bank['loan'])
data_bank['contact'] = label.fit_transform(data_bank['contact'])
data_bank['month'] = label.fit_transform(data_bank['month'])
data_bank['poutcome'] = label.fit_transform(data_bank['poutcome'])
data_bank['y']  = label.fit_transform(data_bank['y'])
```

```python
data_bank
```

```python
X = data_bank.iloc[:,:-1]
y = data_bank.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=12)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
sub_LogModel = LogisticRegression()
sub_LogModel.fit(X_train,y_train)
```

```python
y_pred = sub_LogModel.predict(X_test)
y_pred
```

```python
cv_score = cross_val_score(estimator=sub_LogModel, X=X,y=y,cv=5)
print('cv_score:',cv_score)
print('Mean Accuracy {}\nDeivation {}.'.format(cv_score.mean(),cv_score.std()))
```

```python
from sklearn.metrics import r2_score,accuracy_score,mean_squared_error
```

```python
print("Accuarcy Score :",round(accuracy_score(y_test,y_pred),4))
print("MSE            :",round(mean_squared_error(y_test,y_pred),4))
```
